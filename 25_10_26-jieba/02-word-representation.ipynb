# 1. 使用公开词向量

```python
from gensim.models import KeyedVectors
```

```python
model_path = './data/sgns.weibo.word.bz2'
model = KeyedVectors.load_word2vec_format(model_path)
```

```python
print(model)
```

```python
print(model.vectors)
```

```python
print(model.vector_size)
```

```python
print(model.index_to_key)
# print(model.key_to_index)
```

```python
similarity = model.similarity('地铁', '公交')
print("地铁和公交车的相似度为：", similarity)
```

```python
similar_words = model.most_similar(positive=["上班"])
print(similar_words)
```

```python
result = model.most_similar(positive=["爸爸", "女性"], negative=["男性"], topn=3)
print(result)
```

# 2. 自行训练词向量

```python
import jieba
from gensim.models import Word2Vec, KeyedVectors
import pandas as pd
```

```python
df = pd.read_csv('./data/online_shopping_10_cats.csv', encoding='utf-8', usecols=['review'])
```

```python
sentences = [
    [token for token in jieba.lcut(str(review)) if token.strip() != '']
    for review in df["review"].dropna()
]
```

```python
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python
model.wv.save_word2vec_format('./data/my_word2vec.kv')
```

```python
my_model = KeyedVectors.load_word2vec_format('./data/my_word2vec.kv')
print(my_model)
```

```python
my_model.most_similar(positive=["上班"])
```
